```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_against_date(data_frame_name,x_axis,y_axis,x_label,y_label,plot_title):
    df[x_axis] = pd.to_datetime(df[x_axis])
    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 6))
    ax = sns.lineplot(x=x_axis, y=y_axis, data=data_frame_name)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(plot_title)
    plt.xticks(rotation=45)
    plt.show()

# plot_against_date(df,'Date','Close','Date','JFC Closing Price (in Php)','Historical Price of JFC')
# plot_against_date(df,'Date','Volume','Date','JFC Volume of Traders','Historical count of Traders buying of JFC stocks')
plot_against_date(df,'Date','Close','Date','BTC Closing Price (in Php)','Historical Price of BTC')
plot_against_date(df,'Date','Volume','Date','BTC Volume of Traders','Historical count of Traders buying of BTC stocks')
```
